```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import trim_mean
from scipy.stats import median_abs_deviation

plt.style.use('ggplot')
```

```python
df = pd.read_csv("sugar.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
mean_population = df['Age'].mean()

print("Mean Age =", mean_population)
```

```python
trimmed = trim_mean(df['Age'],0.10)

print("Trimmed Mean =", trimmed)
```

```python
weighted_mean = np.average(
    df['BloodPressure'],
    weights=df['Age']
)

print("Weighted Mean =", weighted_mean)
```

```python
variance = df['Age'].var()

print("Variance =", variance)
```

```python
std = df['Age'].std()

print("Standard Deviation =", std)
```

```python
Q1 = df['Age'].quantile(0.25)

Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1

print("Q1 =",Q1)
print("Q3 =",Q3)
print("IQR =",IQR)
```

```python
print(df['Age'].min())
print(df['Age'].max())
```

```python
print("25th =",df['BloodPressure'].quantile(0.25))

print("50th =",df['BloodPressure'].quantile(0.50))

print("75th =",df['BloodPressure'].quantile(0.75))
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df['Insulin'])

plt.title("Insulin Box Plot")

plt.ylabel("Insulin")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Glucose'],
         bins=10)

plt.title("Glucose Histogram")

plt.xlabel("PGlucose")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['BloodPressure'].plot(kind='density')

plt.title("BloodPressure Density Plot")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Pregnancies'],
            df['Insulin'])

plt.xlabel("Pregnancies")

plt.ylabel("Insulin")

plt.title("Pregnancies vs Insuline")

plt.show()
```

```python
corr = df[['Pregnancies','Age']].corr()

plt.figure(figsize=(5,4))

plt.imshow(corr)

plt.colorbar()

plt.xticks([0,1],
           ['Pregnancies','Age'])

plt.yticks([0,1],
           ['Pregnancies','Age'])

plt.title("Correlation Matrix")

plt.show()
```

```python
import seaborn as sns
```

```python
corr_matrix = df[['Age', 'Pregnancies']].corr()

print(corr_matrix)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=0.5,
            square=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
summary = pd.DataFrame({

'Mean':[df['Age'].mean()],

'Trimmed Mean':[trim_mean(df['Age'],0.10)],

'Median':[df['Age'].median()],

'Std Dev':[df['Age'].std()],

'IQR':[df['Age'].quantile(.75)-df['Age'].quantile(.25)],

'MAD':[median_abs_deviation(df['Age'])]

})

summary
```
