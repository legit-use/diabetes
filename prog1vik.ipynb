```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import trim_mean
from scipy.stats import median_abs_deviation

plt.style.use('ggplot')
```

```python
df = pd.read_csv("Housing.csv")

df.head()
```

```python
mean_area = df['area'].mean()

print("Mean area =", mean_area)
```

```python
trimmed = trim_mean(df['area'],0.10)

print("Trimmed Mean =", trimmed)
```

```python
median_area = df['area'].median()

print("Median =", median_area)
```

```python
weighted_mean = np.average(
    df['area'],
    weights=df['price']
)

print("Weighted Mean =", weighted_mean)
```

```python
variance = df['area'].var()

print("Variance =", variance)
```

```python
std = df['area'].std()

print("Standard Deviation =", std)
```

```python
Q1 = df['price'].quantile(0.25)

Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

print("Q1 =",Q1)
print("Q3 =",Q3)
print("IQR =",IQR)
```

```python
mad = median_abs_deviation(df['price'])

print("MAD =",mad)
```

```python
print(df['price'].min())
print(df['price'].max())
```

```python
print("25th =",df['area'].quantile(0.25))

print("50th =",df['area'].quantile(0.50))

print("75th =",df['area'].quantile(0.75))
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df['area'])

plt.title("area Box Plot")

plt.ylabel("area")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['area'],
         bins=10)

plt.title("area Histogram")

plt.xlabel("area")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['area'].plot(kind='density')

plt.title("area Density Plot")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['area'],
            df['price'])

plt.xlabel("area")

plt.ylabel("price")

plt.title("area vs price")

plt.show()
```
